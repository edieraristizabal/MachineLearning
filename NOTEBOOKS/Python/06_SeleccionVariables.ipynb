## Selección de variables con *Variance Threshold Feature Selection*
Este método es útil para modelos no supervisados, ya que sólo utiliza como criterio la mayor varianza en las variables. Si el valor del umbral es 0 remueve solo las variables de varianza constante. Utilizar un umbral de 0.01 significa eliminar las columnas donde el 99% de los valores son similares.

```python
from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold(threshold=0.25)
selector.fit(X)
print(selector.get_support())
X.columns[selector.get_support()]
```

## Selección de variables univariado

Este método univariable utiliza para seleccionar las variables que tenga la relación mas fuerte con la variable dependiente. Permite seleccionar diferentes *test estadísticos* para seleccionar el número de variables especificado.

Para regresión permite utilizar como *score_func* métodos como *f_regression*, *mutual_info_regression*, y para problemas de clasificación *chi2*, f*_classif*, *mutual_info_classif*. El argumento *k* define el número de variables que deben permanecer eliminando las otras de menor desempeño segun el *score_func*.

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_regression
```

```python
variables = SelectKBest(score_func=mutual_info_regression, k=3)
variables.fit(X,y)
print(variables.get_params)
X.columns[variables.get_support()]
```

## Selección de variables con *Recursive Feature Elimination (RFE)*

En este métodos se utiliza un modelo de *machine learning* para seleccionar las variables, eliminando las de menor importancia en un proceso iterativo. El argumento *step* corresponde al número de variables eliminadas en cada iteración.

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.linear_model import LogisticRegression
rfe=RFE(estimator=LogisticRegression(),n_features_to_select = 5, step = 1)
fit=rfe.fit(X,y)
print(fit.n_features_)
print(fit.support_)
print(fit.ranking_)
```

## Selección de variables con *SelectFromModel*

Este modelo es similar al *RFE*, la diferencia radica en que este está basado en la importancia de los atributos, utilizando como umbral por defecto la media. Como resultado entrega las variables que pasan el umbral.

```python
from sklearn.feature_selection import SelectFromModel
sfm_selector = SelectFromModel(estimator=LogisticRegression())
sfm_selector.fit(X, y)
X.columns[sfm_selector.get_support()]
```

## Selección de variables con *Sequential Feature Selection* (SFS)
Este método utiliza validación cruzada y se desplaza hacia delante (*SFS-Forward*) o atras (*SFS-Backward*) para seleccionar las mejores variables. Inicia con cero variables y encuentra la que mejor optimiza la métrica de la validación cruzada, de esta forma selecciona esta variable y repite el procedimiento para la segunda y demas variables.

```python
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
sfs_selector = SequentialFeatureSelector(estimator=LogisticRegression(), n_features_to_select = 3, cv =10, direction ='backward')
sfs_selector.fit(X, y)
X.columns[sfs_selector.get_support()]
```

## Statistical methods (Two groups - Tests)
During the EDA, we have seen different ways to visualize differences between distributions for landslides and no-landslides. However, we might want to be more rigorous and try to assess the statistical significance of the difference between the distributions, i.e. answer the question “is the observed difference systematic or due to sampling noise?”.

The methodology chosen for hypothesis testing and confidence interval estimation plays a pivotal role in the interpretation of data and subsequent decision-making processes. The bootstrap method, Student's t-test, and the Wilcoxon test are notable among these methodologies. Each offers unique advantages and operates under distinct assumptions, catering to various data types and research questions.

The interpretations will depend on the calculated p-values. Generally, a p-value below a threshold (often 0.05) suggests significant evidence against the null hypothesis, indicating a statistically significant difference between the two groups. 

We are now going to analyze different tests to discern two distributions from each other.


```python
# Group A: More normally distributed
group_A = np.random.normal(100, 20, 100)
# Group B: Slightly skewed distribution
group_B = np.random.normal(105, 25, 100) + np.random.uniform(-10, 10, 100)

# Plot the distributions
sns.histplot(group_A, color='blue', kde=True, label='Group A')
sns.histplot(group_B, color='orange', kde=True, label='Group B')
plt.legend()
plt.title('Distributions of Group A and Group B')
plt.show()
```

* The Bootstrap method. The bootstrap method makes fewer assumptions about the data distribution. It does not assume that the data follows any specific parametric distribution. Methodology: Involves resampling the observed dataset with replacement to create many simulated samples, then calculating the statistic of interest for each sample to form an empirical distribution. This distribution estimates the standard error, constructs confidence intervals, and performs hypothesis testing. The limitations depends on the sample size (larger samples generally provide better approximations). The bootstrap method may not work well with small sample sizes or data with extreme skewness.


```python
def bootstrap_test(group1, group2, n_bootstrap=10000):
    np.random.seed(42)  # Ensuring reproducibility
    diff_means = np.mean(group2) - np.mean(group1)
    combined = np.concatenate([group1, group2])
    count = 0
    for _ in range(n_bootstrap):
        np.random.shuffle(combined)
        new_group1 = combined[:len(group1)]
        new_group2 = combined[len(group1):]
        new_diff_means = np.mean(new_group2) - np.mean(new_group1)
        if new_diff_means >= diff_means:
            count += 1
    p_value = count / n_bootstrap
    return p_value

bootstrap_p_value = bootstrap_test(group_A, group_B)
print(f"Bootstrap p-value: {bootstrap_p_value:.4f}")
```

* T-test. The most common test is the student t-test. T-tests are generally used to compare means. In this case, we want to test whether the means of the income distribution are the same across the two groups. Assumes that the data are normally distributed or that the sample sizes are large enough for the Central Limit Theorem to apply, ensuring that the means are approximately normally distributed. It also assumes homogeneity of variance when comparing two groups (equal variances). The assumptions of normality and equal variances can be restrictive, especially with non-normal data or when variances are unequal.

The test statistic for the two-means comparison test is given by:

$stat=\frac{|\overline{x_1}-\overline{x_2}|}{\sqrt{s^2 / n}}$

Where x̅ is the sample mean and s is the sample standard deviation. Under mild conditions, the test statistic is asymptotically distributed as a Student t distribution.

We use the ttest_ind function from scipy to perform the t-test. The function returns both the test statistic and the implied p-value.

```python
t_stat, t_p_value = stats.ttest_ind(group_A, group_B)
print(f"Student's t-test p-value: {t_p_value:.4f}")
```

The p-value of the test is >0.05, therefore we do not reject the null hypothesis of no difference in means across landslide and no-landslide groups.

The t-test assumes that the variance in the two samples is the same so that its estimate is computed on the joint sample. Welch’s t-test allows for unequal variances in the two samples.

* Standardized Mean Difference (SMD). In general, it is good practice to always perform a test for differences in means on all variables across the treatment and control group, when we are running a randomized control trial or A/B test.

However, since the denominator of the t-test statistic depends on the sample size, the t-test has been criticized for making p-values hard to compare across studies. In fact, we may obtain a significant result in an experiment with a very small magnitude of difference but a large sample size while we may obtain a non-significant result in an experiment with a large magnitude of difference but a small sample size.

One solution that has been proposed is the standardized mean difference (SMD). As the name suggests, this is not a proper test statistic, but just a standardized difference, which can be computed as:

$SMD=\frac{|\overline{x_1}-\overline{x_2}|}{\sqrt{s^{2}_1 + s^{2}_2 / 2}}$

Usually, a value below 0.1 is considered a “small” difference. It is good practice to collect average values of all variables across treatment and control groups and a measure of distance between the two — either the t-test or the SMD — into a table that is called balance table. * Wilcoxon rank-sum test (for independent samples) and the signed-rank test (for paired samples). Non-parametric, does not assume normal distribution of the data. It assumes that the data can be ranked and that the differences between paired observations are symmetrically distributed around the median (for the signed-rank test). Useful for comparing median differences between groups when the data are not normally distributed or when ordinal data are involved. Involves ranking the absolute differences between observations (or the observations themselves in the case of independent samples), then calculating a test statistic based on these ranks. This statistic is used to assess the significance of the observed difference. Generally, it has less power than the t-test for normally distributed data. The interpretation is more about medians than means, which may or may not be preferable depending on the context.

```python
wilcoxon_stat, wilcoxon_p_value = stats.ranksums(group_A, group_B)
print(f"Wilcoxon test p-value: {wilcoxon_p_value:.4f}")
```
