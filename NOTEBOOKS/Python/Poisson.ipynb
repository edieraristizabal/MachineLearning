```python
V = []
Simulations = 10
Trails = 1000
for j in range(0,Simulations):
    f = np.arange(1,Trails+1)
    df = pd.DataFrame(data=f)
    df['Poisson'] = np.random.poisson(Lambda,Trails)
    df.drop([0], axis=1, inplace=True)
    countif = len(df[(df['Poisson'] == 15)])
    count = len(df['Poisson'])
    qq = countif/count
    V.append(qq)
```

In order to predict the default probability, I used the Monte Carlo simulations and the Poisson distribution. The Monte Carlo simulations is widely used in situations where there is no analytical solution and other approaches (such as binomial tree).

The Monte Carlo simulations uses a simulation of risk factors according to an assumed probability distribution and then calculates the default probability for each simulation separately. The average value of the default probability is referred as the estimated value of the default probability.

Standard error of sample mean typically serves as an indication of the level of precision of the model. Since this standard error declines very slow with the number of simulations typically one need to use dozens of thousands (or even more) of sample simulations in order to arrive at an acceptable level of accuracy.

In forecasts, we can determine the probability of occurrence called confidence interval. That is, given two values, what are the chances that the outcome will fall between these two values?

I have used 10 Monte Carlo simulations each of which consists of 1,000 iterations (1000 times of 1 trial) â€” so that in total I have used 10,000 random numbers (10 simulations multiplied by 1,000 iterations) in this calculation.

There is a 90% probability that the final outcome (in this case, the default probability) will be between 6.25% and 8.21%. On the other hand, there is a 5% probability that sales volume will be below 6.25% and another 5% probability that sales volume will be above 8.21%.

That is, the two-tailed confidence interval is a symmetrical interval centered on the mean value, the median value (i.e., the 50th percentile value) or the mode value. Thus, both tails will have the same probability.

The average value of the estimated probability that exactly 15 Israeli companies listed on the Nasdaq will default in the coming year is 7.30%, the median value is 7.35% and the mode value is 7.30%.
